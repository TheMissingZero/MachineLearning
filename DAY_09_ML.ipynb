```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
temp = [20,25,30,35,40]
sale = [13,21,25,35,38]


```

```python
x= np.array([temp]).T # linear regression methos expects a 2d array with rows and cols
print(x)
print(x.shape)
```

```python
y=np.array(sale) # dependent or var that needs to be predicted
print(y)
print(y.shape)
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
print(y_pred)


```

```python
# slope & intercept
print(model.coef_)
print(model.intercept_)
```

```python
# predicting new value
newoutput = model.predict([[45]])
print(newoutput)
```

```python
mse = mean_squared_error(y,y_pred)
print(mse)
```

```python
plt.scatter(temp , sale , marker= "*", color = "r")
plt.plot(temp,y_pred, color='b')
plt.xlabel('temp')
plt.ylabel('sale')
plt.show()
```

```python
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(temp, sale)
temp = [20,25,30,35,40]
sale = [13,21,25,35,38]
print('slope', slope)
print('intercept', intercept)


#user defined fn
def myfunc(x):
  return slope * x + intercept

y_pred = list(map(myfunc,temp))

plt.scatter(temp,sale)
plt.plot(temp,y_pred)
plt.show()


newsale = myfunc(45)
print( newsale )

```
